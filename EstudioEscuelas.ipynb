```python
import pandas as pd
```

```python
schools = pd.read_csv('schools.csv')
```

```python
schools.head(10)
```

## Primer Inciso. 
### Estatus: Resuelto 

```python
best_math_schools = schools[schools['average_math'] > 640][['school_name', 'average_math']].sort_values('average_math',ascending = False)
best_math_schools
```

## Segundo Inciso
### Estatus: Resuelto

```python
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
```

```python
top_10_schools = schools[["school_name", "total_SAT"]].sort_values("total_SAT", ascending = False).head(10)
```

```python
top_10_schools
```

```python
top_10_schools = schools.groupby("school_name", as_index=False)["total_SAT"].mean().sort_values("total_SAT", ascending=False).head(10)
top_10_schools
```

### Tercer Inciso
## estatus: En progreso

#### Numero de escuelas por Distrito

```python
boroughs = schools.groupby("borough")["school_name"].count()
boroughs
```

```python
boroughs = schools.groupby("borough")["total_SAT"].agg(["count", "mean", "std"])
boroughs
```

```python
largest_std_dev = boroughs[boroughs["std"] == boroughs["std"].max()]
largest_std_dev
```

```python
largest_std_dev = largest_std_dev.rename(columns={"count": "num_schools", "mean": "average_SAT", "std": "std_SAT"})
largest_std_dev
```
